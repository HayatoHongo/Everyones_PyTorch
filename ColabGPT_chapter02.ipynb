# 第2章：最急降下法（Gradient Descent）

前回は「平方完成」という数学の力で、一気に谷底（$w=2$）を見つけました。
しかし、現実のAIはもっと複雑で、一気に答えを出すことができません。

そこで、今の場所から**「坂をちょっとずつ下る」**という戦略をとります。これが**最急降下法**です。

---

## 1. 準備：モデルと「勾配」の式

ターゲットにする損失関数 $L$ と、その傾きを表す勾配 $L'$ を用意します。

* **損失関数:** $L = 2.5w^2 - 10w + 10$
* **勾配（傾き）:** $L' = 5w - 10$

> **解説：勾配（こうばい）とは？**
> 「今の場所 $w$ が、どっち向きにどれくらい急な坂になっているか」を表す数値です。

```python
# スタート地点を w=0 とします
w = 0.0

# 学習率（一歩の大きさ）を 0.1 とします
lr = 0.1
```

---

````python
# スタート地点を w=0 とします
w = 0.0

# 学習率（一歩の大きさ）を 0.1 とします
lr = 0.1
````

## 2. 最初の「傾き」を計算する

まずは、今いる場所（$w=0$）の傾きを計算してみましょう。

```python
# 勾配の式 L' = 5w - 10 に w=0 を入れる
grad = 5 * w - 10

print(f"現在の w: {w}")
print(f"この場所の傾き: {grad}")
```

> **解説：傾きが「-10」の意味**
> マイナスの値は「右にいくほど低くなる下り坂」であることを示しています。つまり、$w$ を増やせば誤差が減るということです！

---

````python
# 勾配の式 L' = 5w - 10 に w=0 を入れる
grad = 5 * w - 10

print(f"現在の w: {w}")
print(f"この場所の傾き: {grad}")
````

## 3. 「一歩」だけ進んでみる（更新）

傾きがわかったので、重み $w$ を更新してみましょう。

```python
# 更新のルール： 新しいw = 今のw - (学習率 * 傾き)
w = w - lr * grad

print(f"一歩進んだ後の w: {w}")
```

> **解説：なぜ 1.0 になった？**
> $0 - (0.1 \times -10) = 0 + 1.0 = 1.0$ となりました。
> 谷底の $2.0$ に一歩近づきましたね！

---

````python
# 更新のルール： 新しいw = 今のw - (学習率 * 傾き)
w = w - lr * grad

print(f"一歩進んだ後の w: {w}")
````

## 4. もう一歩だけ進んでみる

さらにもう一度、同じ計算をしてみましょう。

```python
# 今の場所（w=1.0）での傾きを再計算
grad = 5 * w - 10
print(f"新しい場所の傾き: {grad}")

# もう一度更新
w = w - lr * grad
print(f"二歩進んだ後の w: {w}")
```

> **解説：さっきよりも一歩が小さい？**
> 傾きが $-10$ から $-5$ に小さくなりました。
> 谷底に近づくにつれて坂が緩やかになるため、自動的に「慎重に歩く」ようになっています。

---

````python
# 今の場所（w=1.0）での傾きを再計算
grad = 5 * w - 10
print(f"新しい場所の傾き: {grad}")

# もう一度更新
w = w - lr * grad
print(f"二歩進んだ後の w: {w}")
````

## 5. まとめ：学習ループへ

このように「傾きを計算して、ちょっと動く」という作業を繰り返すと、$w$ はどんどん谷底へと吸い寄せられていきます。

この「繰り返し作業」をコンピュータに任せるのが、次のステップである**ループ（for文）**です。

承知いたしました。一歩ずつ、着実に。
「一歩進む」仕組みがわかったところで、それを自動化する「学習ループ」を組み立てていきましょう。

---

## 1. 繰り返しの準備

これまでは手動で計算していましたが、ここからは `for` 文を使ってコンピュータに連続で歩かせます。

まずは $w$ をスタート地点（$0.0$）に戻しましょう。

```python
# パラメータをリセット
w = 0.0
lr = 0.1
```

---

````python
# パラメータをリセット
w = 0.0
lr = 0.1
````

## 2. 学習ループ（10回繰り返す）

「勾配を計算する」→「更新する」という2ステップを10回繰り返してみます。

```python
print("--- 学習スタート ---")

for epoch in range(10):
    # 1. 今の場所での「傾き」を計算
    grad = 5 * w - 10
    
    # 2. 更新（坂を下る）
    w = w - lr * grad
    
    # 3. 結果を表示
    print(f"回数 {epoch+1}: w = {w:.3f}")
```

---

````python
print("--- 学習スタート ---")

for epoch in range(10):
    # 1. 今の場所での「傾き」を計算
    grad = 5 * w - 10

    # 2. 更新（坂を下る）
    w = w - lr * grad

    # 3. 結果を表示
    print(f"回数 {epoch+1}: w = {w:.3f}")
````

## 3. 結果の確認：wはどうなった？

実行結果を見ると、回数を重ねるごとに $w$ が目標の **2.0** に近づいていくのがわかります。

> **解説：学習が進むと歩幅が変わる**
> 1回目は $w=1.0$ まで大きく進みましたが、10回目には $2.0$ にほぼ到達し、動きが止まっています。
> 谷底（傾きが0の地点）に近づくと、更新量（`lr * grad`）も自然に小さくなるからです。

---

## 4. 損失（ズレ）も一緒に計算する

最後に、どれくらい「正解」に近づいたかを数値（損失 $L$）で確認しながらループを回してみましょう。

```python
w = 0.0  # もう一度リセット
lr = 0.1

for epoch in range(10):
    # 傾きの計算
    grad = 5 * w - 10
    # 更新
    w = w - lr * grad
    
    # 今の損失 L = 2.5 * w^2 - 10 * w + 10 を計算
    loss = 2.5 * w**2 - 10 * w + 10
    
    print(f"回数 {epoch+1}: w = {w:.3f}, 損失 = {loss:.6f}")
```

---

````python
w = 0.0  # もう一度リセット
lr = 0.1

for epoch in range(10):
    # 傾きの計算
    grad = 5 * w - 10
    # 更新
    w = w - lr * grad

    # 今の損失 L = 2.5 * w^2 - 10 * w + 10 を計算
    loss = 2.5 * w**2 - 10 * w + 10

    print(f"回数 {epoch+1}: w = {w:.3f}, 損失 = {loss:.6f}")
````

## 5. まとめ

おめでとうございます！これで「機械学習の学習ループ」の基本が完成しました。

* **データの中身を見なくても：** 「勾配（坂の傾き）」さえわかれば、正解にたどり着ける。
* **ループを回す：** コンピュータが少しずつ坂を下ることで、最適な $w$ を自動で見つける。

これが、世界中の高度なAIが裏側で行っていることの「正体」です。
