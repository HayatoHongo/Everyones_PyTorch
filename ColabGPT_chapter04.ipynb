```python
import numpy as np

# 1. データの準備（NumPyの塊をTensorに変換）
import torch
x = torch.tensor([[1.0], [2.0]])
y = torch.tensor([[2.0], [4.0]])


# 重みwをPyTorchの管理下に置く
import torch.nn as nn
model = nn.Linear(1, 1, bias=False)


nn.init.constant_(model.weight, 0.0)

# 道具箱（誤差計算と更新用）
criterion = nn.MSELoss(reduction='mean')

# --- 学習の実行 ---

""" DELETE
# 手順A: 予測して、正解とのズレを測る
y_pred = model.forward(x)            # y_pred = model_linear に相当


loss = criterion.forward(y_pred, y)  # loss = criterion_mse に相当

# 手順B: 勾配を計算する
loss.backward()  # これが all_grads = 2 * (y_pred - y) * x の代わり！a

# 手順C: 実際に重みを動かす
import torch.optim as optim
optimizer = optim.SGD(model.weight, lr=0.1)
optimizer.step()  # w = w - lr * mean_grad を実行

# 結果確認
print(f"PyTorch版 誤差: {loss.item()}")
print(f"PyTorch版 更新後の w: {model.weight.item():.4f}")
"""

########## NEW ##########
epochs = 10

import torch.optim as optim
optimizer = optim.SGD(model.parameters(), lr=0.1)

for epoch in range(epochs):
    # 手順0: 勾配をリセット（前回の勾配が残るのを防ぐ）
    optimizer.zero_grad()

    # 手順A: 予測して、正解とのズレを測る
    y_pred = model.forward(x)            # y_pred = model_linear に相当

    loss = criterion.forward(y_pred, y)  # loss = criterion_mse に相当

    # 手順B: 勾配を計算する
    loss.backward()  # これが all_grads = 2 * (y_pred - y) * x の代わり！a

    # 手順C: 実際に重みを動かす
    optimizer.step()  # w = w - lr * mean_grad を実行

    # 結果確認
    print(f"epoch={epoch+1}/{epochs} | 誤差: {loss.item()} | w: {model.weight.item():.4f}")
########## NEW ##########
```
