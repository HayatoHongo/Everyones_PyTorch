### ライブラリをインポートする

```python
import numpy as np

# 1. データの準備（NumPyの塊をTensorに変換）
""" DELETE
x = np.array([1.0, 2.0])
y = np.array([2.0, 4.0])
"""
########## NEW ##########
import torch
x = torch.tensor([[1.0], [2.0]])
y = torch.tensor([[2.0], [4.0]])
########## NEW ##########

"""DELETE
def model_linear(w, x):
  result = w * x
  return result
"""
########## NEW ##########
# 重みwをPyTorchの管理下に置く
import torch.nn as nn
model = nn.Linear(1, 1, bias=False)
########## NEW ##########

"""DELETE
w = 0.0
"""
########## NEW ##########
nn.init.constant_(model.weight, 0.0)
########## NEW ##########


""" DELETE
def criterion_mse(y_pred, y):
  loss = np.mean((y_pred - y)**2)
  return loss
"""
########## NEW ##########
# 道具箱（誤差計算と更新用）
criterion = nn.MSELoss(reduction='mean')
########## NEW ##########

# --- 学習の実行 ---

# 手順A: 予測して、正解とのズレを測る
""" DELETE
y_pred = model_linear(w, x)
"""
########## NEW ##########
y_pred = model.forward(x)            # y_pred = model_linear に相当
########## NEW ##########

"""DELETE
loss = criterion_mse(y_pred, y)
"""
########## NEW ##########
loss = criterion.forward(y_pred, y)  # loss = criterion_mse に相当
########## NEW ##########

# 手順B: 勾配を計算する
""" DELETE
all_grads = 2 * (y_pred - y) * x
mean_grad = np.mean(all_grads)
"""
########## NEW ##########
loss.backward()  # これが all_grads = 2 * (y_pred - y) * x の代わり！a
########## NEW ##########

# 手順C: 実際に重みを動かす
""" DELETE
lr = 0.1
w = w - lr * mean_grad
"""
########## NEW ##########
import torch.optim as optim
optimizer = optim.SGD(model.weight, lr=0.1)
optimizer.step()  # w = w - lr * mean_grad を実行
########## NEW ##########

# 結果確認
print(f"PyTorch版 誤差: {loss.item()}")
print(f"PyTorch版 更新後の w: {model.weight.item():.4f}")
```
